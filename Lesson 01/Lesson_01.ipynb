### 2.1. Задание (на листочке)
Решите систему уравнений:
$$
\begin{cases}
3 x - 2 y + 5 z = 7\\
7 x + 4 y - 8 z = 3\\
5x - 3 y - 4 z = -12
\end{cases}
$$

Решение:  
Решение производилось путем умножения уравнений на число и суммирование уравнений между собой с целью сократить одну из переменных, а также через выражение одной из оставшихся переменных через другую.  
Ответ:
$$
x = 1 ; y = 3 ; z = 2
$$
Система уравнений линейная.  
Каждое уравнение тоже линейное.

### 4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k = 2
x = np.linspace(-2 * np.pi, 2 * np.pi, 201)
plt.plot(x, np.cos(x), label='cos (x)')
plt.plot(x, np.cos(k * x), label='cos (kx)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
