## Задание 2 (на листочке)  
Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

**Ответ:** Прямые не кажутся перпендикулярными из-за того, что оси X и Y имеют разный масштаб.

## Задание 3 (в программе)  
Напишите код на Python, реализующий построение графиков:
  * окружности,
  * эллипса,
  * гиперболы.

### Окружность

```python
y1 = []
y2 = []
r = 20
x = np.linspace(-r, r, 100)
for i in x:
    y1.append(math.sqrt(r**2 - i**2))
    y2.append(-math.sqrt(r**2 - i**2))
plt.figure(figsize=(4,4))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

### Эллипс

```python
a = 60
b = 30
y1 = []
y2 = []
x = np.linspace(-a, a, 100)
for i in x:
    y1.append(b / a * math.sqrt(a**2 - i**2))
    y2.append(-b / a * math.sqrt(a**2 - i**2))
plt.figure(figsize=(6,3))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

## Задание 4 (на листочке)
  * Пусть задана плоскость: Ax + By + Cz + D = 0  
Напишите уравнение плоскости, параллельной данной и проходящей через начало координат.

**Ответ:** Ax + By + Cz = 0

  * Пусть задана плоскость: A1x + B1y + C1z + D1 = 0 и прямая:  
  $${{x - x_{1}} \over {x_{2} - x_{1}}} = {{y - y_{1}} \over {y_{2} - y_{1}}} = = {{z - z_{1}} \over {z_{2} - z_{1}}}$$

Как узнать, принадлежит прямая плоскости или нет?


**Ответ:** Если выполняется равенство A1x2 + B1y2 + C1z2 + D1 = 0 , то прямая принадлежит плоскости.

## Задание 5 (в программе)
  * Нарисуйте трехмерный график двух параллельных плоскостей.


```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 3*Y
Z2 = 2*X + 3*Y - 15
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z2)
show()
```
