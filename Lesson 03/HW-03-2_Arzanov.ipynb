## Задание 1 (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_func(x, k=1, a=0, b=0):
    return k * np.cos(x - a) + b
```

```python
x = np.linspace(-3*np.pi, 3*np.pi, 201)
plt.plot(x, my_func(x))
plt.plot(x, my_func(x, 0.3))
plt.plot(x, my_func(x, 2, np.pi/2, 1))
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## Задание 3 (в программе)
  * Напишите код, который будет переводить полярные координаты в декартовы.

```python
def to_dec(r=0, alpha=0):
    x = r * np.cos(alpha)
    y = r * np.sin(alpha)
    return x, y
```

```python
alpha = np.linspace(0, 2*np.pi, 100)
r = 40
x, y = to_dec(r, alpha)
plt.figure(figsize=(4,4))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## Задание 4 (в программе)
Решите систему уравнений:  
$$ \begin{cases}
y = x^2 – 1\\
\exp(x) + x \cdot (1 – y) = 1
\end{cases}  
$$


```python
x = np.linspace(-5, 5, 201)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,8) 
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)

x1, y1 = fsolve(equations, (-2, 2))
x2, y2 = fsolve(equations, (2, 5))            
print (x1, y1)
print (x2, y2)
```
